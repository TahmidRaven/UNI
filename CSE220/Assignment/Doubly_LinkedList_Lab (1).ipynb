```python
class Node:
  def __init__(self, e, n, p):
    self.element = e
    self.next = n
    self.prev = p
```

```python
class DoublyList:
  
  def __init__(self, a):
  # Creates a Non Dummy Headed Circular Doubly Linked List using the values from the given array a.
    # To Do
    # pass # Remove 
    
    self.head = Node(a[0], None, None)
    tail = self.head
    for i in range(1, len(a)):
        newNode = Node(a[i], None, None)
        tail.next = newNode
        newNode.prev = tail
        tail = newNode
    self.head.prev = tail
    tail.next = self.head
    

  
  # Counts the number of Nodes in the list and return the number
  # def countNode(self):
    # To Do
    # pass # Remove this line
  def countNode(self):
    counter = 1
    head = self.head
    n = head.next
    while n != head:
        counter += 1
        n = n.next
    return counter
  
  
  # prints the elements in the list
  def forwardprint(self):
    # To Do
    # pass # Remove this line
    cur = self.head
    head = self.head
    n = head.next
    while n != head:
        print(cur.element, end = ", ")
        cur = n
        n = n.next
    print(cur.element)

  # prints the elements in the list backward
  def backwardprint(self):
    # To Do
    # pass # Remove this line
    head = self.head
    n = head.next
    last = None
    
    while n != head:
        last = n
        n = n.next
    
    while last != head:
        print(last.element, end = ", ")
        last = last.prev
    print(last.element)

  # returns the reference of the at the given index. For invalid index return None.
  def nodeAt(self, idx):
    # To Do
    # pass # Remove this 
    if idx >= self.countNode():
        print("index error")
    else:
        counter = 0
        n = self.head
        while counter != idx:
            counter += 1
            n = n.next
        return n
      
  
  # returns the index of the containing the given element. if the element does not exist in the List, return -1.
  def indexOf(self, elem):
    # To Do
    # pass # Remove this line

    counter = 0
    n = self.head
    while n.element != elem:
        if counter == self.countNode():
            return -1
        counter += 1
        n = n.next
    return counter


  # inserts containing the given element at the given index Check validity of index. 
  def insert(self, elem, idx):
    # To Do
    # pass # Remove this line

    newNode = Node(elem, None, None)
    if idx == 0:
        prev = self.head.prev
        newNode.next = self.head
        self.head.prev = newNode
        self.head = newNode
        self.head.prev = prev
        prev.next = self.head
    else:
        prex = self.nodeAt(idx-1)
        newNode.next = prex.next
        prex.next.prev = newNode
        prex.next = newNode
        newNode.prev = prex
    
  # removes at the given index. returns element of the removed node. Check validity of index. return None if index is invalid.
  def remove(self, idx):
    # To Do
    # pass # Remove this line
    if idx == 0:
        removed = self.head
        self.head.prev.next = self.head.next
        self.head.next.prev = self.head
        self.head = self.head.next
        return f"{removed.element}"
    else:
        pred = self.nodeAt(idx-1)
        removed = pred.next
        pred.next = removed.next
        removed.next.prev = pred
        return f"{removed.element}"
    
  # def remove(self, idx):
  #     if idx >= self.countNode():
  #         print("index error")
  #         return None
  #     else:
  #         node_to_remove = self.nodeAt(idx)
  #         node_to_remove.prev.next = node_to_remove.next
  #         node_to_remove.next.prev = node_to_remove.prev
  #         return f"{node_to_remove.element}"

  
      
    
```
